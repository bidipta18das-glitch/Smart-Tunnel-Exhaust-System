```python
import pandas as pd

df = pd.read_csv("../data/raw/tunnel_ventilation.csv")

df = df.drop(columns=["timestamp"])

df = df.rename(columns={
    "temperature": "Temperature",
    "humidity": "Humidity",
    "smoke": "MQ2_Raw",
    "alcohol": "MQ3_Raw"
})

df.to_csv("../data/raw/tunnel_ventilation.csv", index=False)

print("Dataset updated successfully")
print(df.head())
```
